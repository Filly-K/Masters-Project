```python
import numpy as np
import pandas as pd
import scipy as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
test_data = pd.read_csv('pre_post.csv')
test_data 
```

```python
from scipy.stats import shapiro

results = []
for c in test_data.columns:
    #print(c)
    s = test_data[c].dropna()
    stat, p = shapiro(s)
    interpret = 'Normal' if p>0.05 else "Not normal"
    results.append((c, stat, p, interpret))
#print(results)
#for col, stat, p, interpret in results:
    #print(col, 'stat=%.3f, p=%.3f\n' % (stat, p))
df_shapiro = pd.DataFrame(results, columns=['column', 'stat', 'p', 'interpret'])
df_shapiro.to_csv('pre_post_shapiro.csv', index=False)
df_shapiro
```
